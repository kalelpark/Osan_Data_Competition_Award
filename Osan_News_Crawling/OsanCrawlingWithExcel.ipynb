<h1> 오산시 불법주차 관련 뉴스 크롤링(Crawling) </h1>

```python
import requests
from bs4 import BeautifulSoup
import openpyxl
import json
import pprint

# NAVER Developer Login
client_id = 'sUFtTNfETjYapyKveAFM'
client_pw = 'CVpbS5_674'

Osan_news_crawling = list()

# Using RestAPI
num = 0
start_num = 1
for index in range(10):
  start_number = start_num + (index*100)
  naver_open_api = 'https://openapi.naver.com/v1/search/news.json?query=오산시 불법주차&display=100&start='+str(start_number)
  header_params = {'X-Naver-Client-id': client_id , 'X-Naver-Client-Secret': client_pw}
  res = requests.get(naver_open_api, headers = header_params)

  if res.status_code == 200:
    data = res.json()
    for item in data['items']:
      Osan_news_crawling.append([item['title'], item['originallink']])
      
# 출력하기
for Osan in Osan_news_crawling:
  print('제목 :', Osan[0])
  print('원본 주소 :', Osan[1])
  print()
```

## 오산시 불법주정차 관련 뉴스 Excel 저장 ##

```python
# NAVER API Login
client_id = 'sUFtTNfETjYapyKveAFM'
client_pw = 'CVpbS5_674'

# OPEN Excel
excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['B'].width = 100
excel_sheet.column_dimensions['C'].width = 100
excel_sheet.append(['날짜', '제목', '링크'])

num = 0
start_num = 1
for index in range(10):
  start_number = start_num + (index*100)
  naver_open_api = 'https://openapi.naver.com/v1/search/news.json?query=오산시 불법주차&display=100&start=' + str(start_number)
  header_params = {'X-Naver-Client-id': client_id , 'X-Naver-Client-Secret': client_pw}

  res = requests.get(naver_open_api, headers = header_params)

  if res.status_code == 200:
    # Switch -> json File
    data = res.json()
    for item in data['items']:
      num += 1
      excel_sheet.append([item['pubDate'][11:] , item['title'], item['originallink']])
excel_file.save('Osan_news.xlsx')
excel_file.close()
```
